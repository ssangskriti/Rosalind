# 3A Generate the k-mer Composition of a String

```python
n = int(input())
string = input()

listy = []

for i in range(0,len(string)):
  sub = string[i:i+n]
  if (len(sub)==n):
    listy.append(sub)
  
listy.sort()

for i in listy:
  print (i)
```

#  3B Reconstruct a String from its Genome Path

```python
string = input()
listy = string.split()
def recon(listy):
    string = listy[0]
    for i in range(1, len(listy)):
        string += listy[i][len(listy[i])-1]
    return string

print(recon(listy))
    
```

# 3C Construct the Overlap Graph of a Collection of k-mers

```python
string = input()
listy = string.split()

for i in range(0,len(listy)):
  suffix = listy[i][0:len(listy[i])-1]
  
  for j in range (0 , len(listy)):
    if i != j:
      prefix = listy[j][1:len(listy[i])]
      
      if suffix == prefix: 
        print (listy[j], "-> ", end="")
        print (listy[i])
```

# 3D Construct the De Bruijn Graph of a String

```python
k = int(input())
string = input()
listy = {}
k = k-1
for i in range(0, len(string)-k):
    if(string[i:i+k] in listy):
        listy[string[i:i+k]].append(string[i+1:i+k+1])
    else:
        listy[string[i:i+k]] = []
        listy[string[i:i+k]].append(string[i+1:i+k+1])
    
for key, value in listy.items():
    print(key, '->', ','.join(value))
```

```python
string = input()
listy = string.split()
di = {}
for i in listy:
    s = i[0:len(i)-1]
    p = i[1:len(i)]
    if s in di:
        di[s].append(p)
    else:
        di[s] = []
        di[s].append(p)
for key, value in di.items():
    print(key, '->', ','.join(value))
```
