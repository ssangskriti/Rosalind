# Generate the k-mer Composition of a String

```python
n = int(input())
string = input()

listy = []

for i in range(0,len(string)):
  sub = string[i:i+n]
  if (len(sub)==n):
    listy.append(sub)
  
listy.sort()

for i in listy:
  print (i)
```

# Construct the Overlap Graph of a Collection of k-mers

```python
string = input()
listy = string.split()

for i in range(0,len(listy)):
  suffix = listy[i][0:len(listy[i])-1]
  
  for j in range (0 , len(listy)):
    if i != j:
      prefix = listy[j][1:len(listy[i])]
      
      if suffix == prefix: 
        print (listy[j], "-> ", end="")
        print (listy[i])
```
