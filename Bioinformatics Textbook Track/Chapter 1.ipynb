# Find a Position in a Genome Minimizing the Skew

```python
genome = ''

def skew(genome):
    c = 0
    g = 0
  
    min_skew = 0
    pos_list = []
    position = 0
  
    for i in genome:
        position += 1
    
        if i == 'G':
            g += 1
        elif i == 'C':
            c += 1
      
        skew = g-c
    
        if skew < min_skew:
            pos_list = [position]
            min_skew = skew
        elif skew == min_skew and position not in pos_list:
            pos_list.append(position)	
      
    print(pos_list)

skew(genome)       
```
