# 1C Find the Reverse Complement of a String

```python
string = input()
def reverse_compliment(string):
    p = []
    for i in string:
        if i == 'A':
            p.append('T')
        elif i == 'T':
            p.append('A')
        elif i == 'G':
            p.append('C')
        elif i == 'C':
            p.append('G')
    s = []
    for i in range(len(p)-1, -1, -1):
        s.append(p[i])
    for i in s:
        print(i,end="")
    
reverse_compliment(string)        
```

# 1E Find Patterns Forming Clumps in a String

```python
string = input()
k = input()
k = int(k)
l = input()
l = int(l)
t = input()
t = int(t)
def clumps(string, k, l, t):
    finalListy = []
    for i in range(0, (len(string)-l)):
        s = string[i:i+l]
        listy = {}
        for j in range(0, len(s)-k):
            p = s[j:j+k]
            if p in listy:
                listy[p] += 1
            else:
                listy[p] = 1
        for key, value in listy.items():
            if value >= t:
                finalListy.append(key)
    listy = []
    for i in finalListy:
        if i in listy:
            continue;
        else:
            print(i, end=" ")
            listy.append(i)
clumps(string, k, l, t)
```

# 1F Find a Position in a Genome Minimizing the Skew

```python
genome = ''

def skew(genome):
    c = 0
    g = 0
  
    min_skew = 0
    pos_list = []
    position = 0
  
    for i in genome:
        position += 1
    
        if i == 'G':
            g += 1
        elif i == 'C':
            c += 1
      
        skew = g-c
    
        if skew < min_skew:
            pos_list = [position]
            min_skew = skew
        elif skew == min_skew and position not in pos_list:
            pos_list.append(position)	
      
    print(pos_list)

skew(genome)       
```

# 1G Compute the Hamming Distance Between Two Strings

```python
p = input()
q = input()
def hamming_dist(p, q):
    count = 0
    for (i,j) in zip(p,q):
        if i != j:
            count += 1
    return count
print(hamming_dist(p,q))
```

# 1H Find All Approximate Occurrences of a Pattern in a String

```python
pattern = input()
string = input()
d = int(input())
def occur(pattern, string, d):
    listy = []
    for i in range(0, len(string)-len(pattern)):
        s = string[i:i+len(pattern)]
        diff = 0
        for (j,k) in zip(s, pattern):
            if j!= k:
                diff += 1
            if diff > d:
                break
        if diff <= d and k == pattern[len(pattern) -1 ]:
            listy.append(i)
    return listy
listy = occur(pattern, string, d)
for i in listy:
    print(i, end=" ")
```

# 1L Implement PatternToNumber

```python
dna = input()
def patternToNumber(dna):
    d = {'A':0, 'C':1, 'G':2, 'T':3}
    l = len(dna)-1
    n = 0
    for i in dna:
        s = d[i] * pow(4,l)
        n += s
        l -= 1
    return n

print(patternToNumber(dna))
```

# 1N Generate the d-Neighborhood of a String
