#BA4B	Find Substrings of a Genome Encoding a Given Amino Acid String

```
string = input()
pattern = input()

rnaCodonToDna = {
    'TTT': 'F',     'CTT': 'L',     'ATT': 'I',     'GTT': 'V',
    'TTC': 'F',     'CTC': 'L',     'ATC': 'I',     'GTC': 'V',
    'TTA': 'L',     'CTA': 'L',     'ATA': 'I',     'GTA': 'V',
    'TTG': 'L',     'CTG': 'L',     'ATG': 'M',     'GTG': 'V',
    'TCT': 'S',     'CCT': 'P',     'ACT': 'T',     'GCT': 'A',
    'TCC': 'S',     'CCC': 'P',     'ACC': 'T',     'GCC': 'A',
    'TCA': 'S',     'CCA': 'P',     'ACA': 'T',     'GCA': 'A',
    'TCG': 'S',     'CCG': 'P',     'ACG': 'T',     'GCG': 'A',
    'TAT': 'Y',     'CAT': 'H',     'AAT': 'N',     'GAT': 'D',
    'TAC': 'Y',     'CAC': 'H',     'AAC': 'N',     'GAC': 'D',
    'TAA': 'Stop',  'CAA': 'Q',     'AAA': 'K',     'GAA': 'E',
    'TAG': 'Stop',  'CAG': 'Q',     'AAG': 'K',     'GAG': 'E',
    'TGT': 'C',     'CGT': 'R',     'AGT': 'S',     'GGT': 'G',
    'TGC': 'C',     'CGC': 'R',     'AGC': 'S',     'GGC': 'G',
    'TGA': 'Stop',  'CGA': 'R',     'AGA': 'R',     'GGA': 'G',
    'TGG': 'W',     'CGG': 'R',     'AGG': 'R',     'GGG': 'G'
}

def reversecompliment (string):
  output = ""
  for i in range(len(string)-1, -1, -1):
    if (string[i] == "A"):
      output += "T"
    elif (string[i] == "T"):
      output += "A"
    elif (string[i] == "G"):
      output += "C"
    elif (string[i] == "C"):
      output += "G"
  return output

listy = []

for i in range(0,len(string)-(3*len(pattern))+1):
  s = string[i:i+(3*len(pattern))]
  temp = 0
  k = 0
  for j in range(0, len(pattern)):
    if rnaCodonToDna[s[k:k+3]] == pattern[j] or rnaCodonToDna[reversecompliment(s[k:k+3])] == pattern[len(pattern)-1-j]:
      temp += 1
    k += 3
  if temp ==len(pattern):
    listy.append(s)

for i in range(0,len(listy)):
  print (listy[i])
```

#BA4C Generate the Theoretical Spectrum of a Cyclic Peptide

```
mass = {}
mass['G'] = 57
mass['A'] = 71
mass['S'] = 87
mass['P'] = 97
mass['V'] = 99
mass['T'] = 101
mass['C'] = 103
mass['I'] = 113
mass['L'] = 113
mass['N'] = 114
mass['D'] = 115
mass['K'] = 128
mass['Q'] = 128
mass['E'] = 129
mass['M'] = 131
mass['H'] = 137
mass['F'] = 147
mass['R'] = 156
mass['Y'] = 163
mass['W'] = 186

def CycSpectrum(dna):

  def maass(s):
    listy = 0
    for i in s:
      listy+=mass[i]
    return listy

  length = len(dna)
  dna=dna*2
  listy = []
  for j in range(1,length):
    for i in range(0,length):
      listy.append(dna[i:i+j])
  
  return sorted([0]+[maass(s) for s in listy]+[maass(dna[:length])])

print(CycSpectrum("HEDWKPAMCVIEGQ"))
```

#BA4D Compute the Number of Peptides of Given Total Mass

```
keys = [57, 71, 87, 97, 99, 101, 103, 113, 114, 115, 128, 129, 131, 137, 147, 156, 163, 186]
def var(s):
    listy = [0]*(s + 1)
    index = s
    listy[s] = 1
    while index > 0:
        for key in keys:
            listy[index-key] += listy[index]

        index -= 1
        while listy[index] == 0:
            index -= 1
    return listy[0]
n = 1024
print(var(n))
```
